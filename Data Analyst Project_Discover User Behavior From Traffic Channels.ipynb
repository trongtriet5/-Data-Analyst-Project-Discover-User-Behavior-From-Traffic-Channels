**Phân tích dữ liệu hành vi người dùng cho ngành Tài chính, Chứng khoán tại Pinetree**

Đề tài này nhằm phân tích và đánh giá các chỉ số hành vi của người dùng đến từ nhiều nguồn khác nhau trên trang web của Pinetree, một công ty trong ngành tài chính và chứng khoán. Cụ thể, chúng tôi sẽ xem xét các chỉ số liên quan đến traffic, nguồn gốc, sự kiện và chuyển đổi để đánh giá chất lượng và hiệu quả của các nguồn traffic khác nhau. Mục tiêu của phân tích này bao gồm:

1. **Đánh giá chất lượng traffic, nguồn gốc và sự kiện:** Sử dụng các tiêu chí đánh giá cụ thể để so sánh và đưa ra bảng số liệu chi tiết về các nguồn traffic, sự kiện và tỷ lệ chuyển đổi từ các nguồn này.
2. **Đề xuất chiến lược tăng trưởng traffic:** Dựa trên phân tích tỷ lệ chuyển đổi và hiệu quả của các nguồn traffic hiện tại, chúng tôi sẽ đề xuất các nguồn đầu tư phù hợp để vừa tăng số lượng traffic vừa đảm bảo tỷ lệ chuyển đổi tốt.
3. **Phân tích biến động hành vi người dùng theo thời gian:** Đánh giá biến động của các chỉ số hành vi người dùng và sự kiện theo từng tuần để xác định những hành vi quan trọng nhất, từ đó rút ra các kết luận và bài học quan trọng.
4. **Đề xuất cải thiện landing page:** So sánh landing page của Pinetree với các đối thủ chính trong ngành (Chứng khoán Shinhan, VPS, SSI, XTB), đề xuất các thay đổi cần thiết để cải thiện trải nghiệm người dùng và hỗ trợ tốt hơn cho mục tiêu tiếp thị của Pinetree.

Đối tượng mục tiêu của Pinetree bao gồm hai nhóm chính:
1. Nhà đầu tư mới (18-35 tuổi) sống ở các thành phố lớn và quan tâm đến đầu tư tài chính, chứng khoán.
2. Nhà đầu tư chuyên nghiệp (25+ tuổi) đã mở tài khoản và có kinh nghiệm trading lớn, sử dụng chứng khoán như kênh thu nhập chính.

Mục tiêu tiếp thị của Pinetree là tiếp cận và khuyến khích nhóm nhà đầu tư mới mở tài khoản chứng khoán tại Pinetree và thúc đẩy giao dịch từ nhóm nhà đầu tư chuyên nghiệp thông qua các ưu đãi phí giao dịch.

Phân tích này sẽ sử dụng các công cụ và nguồn dữ liệu như Speed Test, Website Similar, và các công cụ nghiên cứu từ khóa (keywordtool.io, Google Keyword Planner) để hỗ trợ việc đánh giá và đưa ra các đề xuất cải tiến phù hợp.

- `Session source / medium`: Nguồn và phương tiện của phiên (ví dụ: Google/Organic, Facebook/Ads).
- `Active users`: Số lượng người dùng hoạt động trong một khoảng thời gian cụ thể.
- `Sessions`: Tổng số phiên người dùng truy cập vào trang web.
- `Engaged sessions`: Số phiên mà người dùng tương tác với trang web (ví dụ: thời gian ở lại lâu, nhiều trang được xem).

**Dựa trên ma trận tương quan ta có thể thấy thêm một số mối tương quan của các đặc trựng với nhau . Để nhìn rõ hơn, ta sẽ filter lại ma trận tương quan để có thể thấy rõ hơn về mối tương quan của nó**

```python
# Tính toán ma trận tương quan
correlation_matrix = df_weekly[['Active users', 'Sessions', 'Key events',
                                'Engaged sessions']].corr()

# Lọc ma trận tương quan
threshold = 0.9
filtered_corr_matrix = correlation_matrix.copy()
filtered_corr_matrix[(filtered_corr_matrix < threshold) & (filtered_corr_matrix > -threshold)] = None

# Thiết lập kích thước hình vẽ
plt.figure(figsize=(10, 8))

# Vẽ heatmap
sns.heatmap(filtered_corr_matrix, annot=True, cmap='coolwarm', fmt='.2f',
            linewidths=0.5, linecolor='black', vmin=-1, vmax=1, center=0,
            mask=filtered_corr_matrix.isna())

# Thiết lập tiêu đề và nhãn
plt.title('Correlation Matrix Filterd (|r| > 0.9)', fontsize=16)
plt.show()
```

- **Tăng Cường Tương Tác**: Xem xét các sự kiện có tỷ lệ thấp hơn và tìm cách cải thiện các điểm tương tác này. Ví dụ, tối ưu hóa các nút hoặc biểu mẫu có thể giúp tăng tỷ lệ `form_start` và `form_submit`.
- **Phân Tích Sâu Hơn**: Kiểm tra các yếu tố cụ thể ảnh hưởng đến các sự kiện có tỷ lệ tương tác cao và thấp để hiểu rõ hơn về hành vi của người dùng.

```python
# Tính toán tổng số sự kiện và sessions cho từng Event name và Session source / medium
top_events = df_event_details.groupby(['Event name', 'Session source / medium']).agg({
    'Event count': 'sum',
    'Sessions': 'sum'
}).reset_index()

# Sắp xếp các sự kiện theo giá trị giảm dần của Event count
sorted_events = top_events.sort_values(by='Event count', ascending=False).reset_index(drop=True)

sorted_events.head(10)
```

```python
# Gộp các giá trị từ vị trí 101 trở lên thành "Others/medium"
top_100_events = sorted_events.head(100)
others = sorted_events.iloc[100:].copy()
others['Event name'] = 'Others'
others['Session source / medium'] = 'Others'

# Gộp lại dataframe
final_events = pd.concat([top_100_events, others.groupby(['Event name', 'Session source / medium']).sum().reset_index()])

print(final_events)
```

```python
# Tạo biểu đồ bar cho kết quả cuối cùng
plt.figure(figsize=(14, 10))
barplot = sns.barplot(data=final_events, x='Event count', y='Event name', hue='Session source / medium', dodge=False)
plt.title('Top 100 Events by Event Count (Others Grouped)')
plt.xlabel('Event Count')
plt.ylabel('Event Name')
plt.legend(title='Session source / medium', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# In chi tiết về các giá trị trên biểu đồ
for i, row in final_events.iterrows():
    print(f"Event: {row['Event name']} | Source/Medium: {row['Session source / medium']} | Event Count: {row['Event count']} | Sessions: {row['Sessions']}")
```

```python
# Lấy danh sách các Event name duy nhất
unique_events = final_events['Event name'].unique()

# Tạo biểu đồ cho từng Event name
for event in unique_events:
    event_data = final_events[final_events['Event name'] == event]

    plt.figure(figsize=(14, 10))
    barplot = sns.barplot(data=event_data, x='Event count', y='Session source / medium', dodge=False)
    plt.title(f'Event Count for {event}')
    plt.xlabel('Event Count')
    plt.ylabel('Session source / medium')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    # Thêm các ghi chú hiển thị chi tiết hơn về các giá trị trên biểu đồ
    for p in barplot.patches:
        width = p.get_width()
        plt.text(
            width + 1,
            p.get_y() + p.get_height() / 2,
            f'{width:.0f}',
            ha='left',
            va='center'
        )

    plt.show()
```

**1. Tăng Cường Đề Xuất Giá Trị (Value Proposition)**

**1.1 Đề Xuất:**

**1.4 Giải Pháp**

**1.4.1 Xác Định Vấn Đề và Nhu Cầu Của Khách Hàng**
- **Mục Tiêu:** Hiểu rõ các vấn đề và nhu cầu của khách hàng để xây dựng Value Proposition phù hợp.
- **Cách Thực Hiện:** Nghiên cứu thị trường qua khảo sát và phỏng vấn. Từ đó, tạo Value Proposition nhằm giải quyết vấn đề và đáp ứng nhu cầu khách hàng. Chú ý đến các từ khóa khách hàng hay nhắc đến để tăng tính kết nối.

**1.4.2 Tuyên Bố Lợi Ích Một Cách Rõ Ràng**
- **Mục Tiêu:** Trình bày Value Proposition rõ ràng, giải quyết vấn đề và thỏa mãn mong muốn của khách hàng.
- **Lời Khuyên:** Value Proposition nên ngắn gọn (2-3 câu), rõ ràng và mạch lạc. Điều này giúp khách hàng nhận thấy giá trị mà họ sẽ nhận được từ sản phẩm/dịch vụ.

**1.4.3 Tập Trung Mang Lại Giá Trị Cho Khách Hàng**

**1.4.4 Tạo Sự Khác Biệt So Với Đối Thủ**

- **Cách Thực Hiện:** Nghiên cứu đối thủ để hiểu điểm mạnh và yếu của họ. Điều chỉnh Value Proposition để mang lại giá trị độc đáo mà đối thủ không có.
